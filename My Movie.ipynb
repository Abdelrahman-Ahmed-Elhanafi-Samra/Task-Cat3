```python
movie = movie.drop(movie.index[1105])
movie.reset_index(drop=True)
```

```python
pd.to_numeric([movie['Vote_Count'],movie['Vote_Average']], errors='coerce')
```

```python
movie['Vote_Count'] = movie['Vote_Count'].astype('float')
movie['Vote_Average'] = movie['Vote_Average'].astype('float')
```

```python
movie['Original_Language'].unique()
```

```python
a=movie['Genre'].unique()
```

```python
from fuzzywuzzy import process
```

```python
basic_genres = pd.Series(['Action','Comedy','Drama','Fantasy','Horror','Mystery','Romance','Thriller','Western'])
for state in basic_genres:
    matches = process.extract(state, movie['Genre'], limit = movie.shape[0])
    for potential_match in matches:
        if potential_match[1] >= 60:
            movie.loc[movie['Genre'] == potential_match[0], 'Genre'] = state
```

```python
movie['Genre'].unique()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=movie, y='Genre', color='blue', order= movie['Genre'].value_counts().index[:12])
plt.title('Most genre of movies')
```

* From the graphic representation here, it appears that action is the most produced genre of filmsFrom the graphic representation here, it appears that action is the most produced genre of films

```python
sns.catplot(x='Vote_Count', y='Title', data=movie, kind='box', order= movie['Title'].value_counts().index[:12])
plt.title('Movie reviews')
```

* From the graph here, it appears that Beauty and the Beast is the most rated movie

```python
movie['year'] = pd.DatetimeIndex(movie['Release_Date']).year
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=movie, y='year', color='red', order= movie['year'].value_counts().index[:12])
plt.title('Most Release_Date of movies')
```
